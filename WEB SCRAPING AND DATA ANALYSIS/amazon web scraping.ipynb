```python
from bs4 import BeautifulSoup
import pandas as np
import requests
# url = 'https://www.amazon.com/Show-Data-T-Shirt-BCBA-Analyst/dp/B0CD4HMM8P/ref=sr_1_7?crid=3K8R3WS19T1Z&keywords=data+analyst+t+shirt&qid=1707449820&sprefix=data+analysit+tsi%2Caps%2C588&sr=8-7'
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_films'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
soup.find('table', class_ ="wikitable sortable plainrowheaders sticky-header col4right col5center col6center")
```

```python
table = soup.find_all('table')[1]
table
```

```python
soup.find('table')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
[title.text for title in world_titles]
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
world_table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```
