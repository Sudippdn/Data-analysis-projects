### Filtering Data

a. Finding particular data in a dataset

```python
Finding_Mega = df.loc[df['Name'].str.contains("Mega")]
Finding_Mega.head(5)
```

b. Removing the data in a dataset

```python
Removing_Mega = df.loc[~df['Name'].str.contains("Mega")]
Removing_Mega.tail(5)
```

c. Finding particular string words in a dataset

1. If you want to get only mega then do as (a) 
2. but if you introduce "flags = re.I" after "import re" then string can have either capital or small letters doesn't matter

```python
import re
String_Finder = df.loc[df["Type 1"].str.contains("FirE|GRass|mega", flags = re.I, regex= True)]
String_Finder.head()
```

### Conditional Change
