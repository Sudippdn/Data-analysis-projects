# Visualization in Pandas

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('seaborn-v0_8-ticks')
```

### Creating a line plot

df.plot() this is default to line plot

```python
df.plot(kind = 'line', title = "Ice Cream Ratings", xlabel = "Daily Rating", ylabel = "Scores")
```

### For vertical Bar plot

```python
df.plot(kind = 'bar')
```

### for horizontal bar plot

```python
df.plot.barh(stacked = True)
```

### Scatter plot

```python
df.plot.scatter(x = 'Texture Rating', y= 'Overall Rating', s= 500, c = 'red')
```

### Histogram plot

```python
df.plot.hist(bins = 15)
```

#### Box Plot

```python
df.boxplot()
```

### Area Plot

```python
df.plot.pie(y = "Flavor Rating")
```

### Area Plot

```python
df.plot.area(figsize = (10,5))
```
